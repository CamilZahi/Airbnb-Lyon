```python
# Importation des librairies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium import plugins  
```

---
<h3><center> Les données </center></h3>

```python
# Importation des données
df = pd.read_csv('AIRBNB LYON net.csv', sep=',')
```

---
<h3><center> Moteur de recherche avec vue géographique </center></h3>

```python
# Paramètres de la figure de base
fig = folium.Map(location=[45.759060, 4.847331], zoom_start=14)

# Configuration d'arrière plan de la carte
folium.TileLayer('Stamen Terrain').add_to(fig)
folium.LayerControl().add_to(fig)
# Option qui permet de dézoomer
minimap = plugins.MiniMap(toggle_display=True).add_to(fig)
# Ajout des logements disponibles à la carte
points = list(zip(df['Latitude'],df['Longitude']))
prixparnuits = list(df["Prix par nuit"])
localisation = list(df["Localisation"])
typelogements = list(df["Type"])
nbinvités = list(df["Nombre d'invités"])
nbchambres = list(df["Nombre de chambres"])
nblits = list(df["Nombre de lits"])
notemoyenne = list(df["Note moyenne sur 5"])
a=0
b=0
c=0
d=0
e=0
f=0
g=0
print("Veuillez entrer seulement des chiffres")
prix_min = int(input("Prix minimum : "))
prix_max = int(input("Prix maximum : "))
arrondissement = int(input("Arrondissement (de 1 à 9)"))
nb_invités_min = int(input("Nombre d'invités minimum : "))
nb_invités_max = int(input("Nombre d'invités maximum : "))
nb_chambres_min = int(input("Nombre de chambres minimum : "))
nb_chambres_max = int(input("Nombre chambres maximum : "))
nb_lits_min = int(input("Nombre de lits minimum : "))
nb_lits_max = int(input("Nombre lits maximum : "))
note_minimale = float(input("Note minimale : "))
for p in points:
    if ((prix_min <= prixparnuits[b] <= prix_max) and (localisation[c] == arrondissement) and (nb_invités_min <= nbinvités[d] <= nb_invités_max) and (nb_chambres_min <= nbchambres[e] <= nb_chambres_max) and (nb_lits_min <= nblits[f] <= nb_lits_max) and (notemoyenne[g] >= note_minimale)):
        pt = folium.Marker(location=points[a], tooltip=prixparnuits[b], icon=folium.Icon(color='green', icon='home', prefix = 'glyphicon'))
        pt.add_to(fig)          
    a=a+1
    b=b+1
    c=c+1
    d=d+1
    e=e+1
    f=f+1
    g=g+1
fig
```
